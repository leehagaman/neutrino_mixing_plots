```python
import matplotlib.pyplot as plt
import numpy as np

import nufit_params

delta_m21_sq = nufit_params.delta2_m21  # Solar mass splitting (eV^2)
delta_m31_sq = nufit_params.delta2_m3l  # Atmospheric mass splitting (eV^2)
```

```python
plt.rcParams.update({'font.size': 14})

x_values = np.array([1, 1e8])

plt.figure(figsize=(10, 8))
ax = plt.gca()

ax.plot(x_values, delta_m21_sq * x_values, 'k--', linewidth=2)
ax.text(1e5, 15, r'$\Delta m_{21}^2$-driven first maximum', rotation=47, fontsize=12)
ax.plot(x_values, delta_m31_sq * x_values, 'k--', linewidth=2)
ax.text(3e3, 15, r'$\Delta m_{31}^2$-driven first maximum', rotation=47, fontsize=12)

# Rectangles and labels
def add_box(xmin, xmax, ymin, ymax, color, text, text_pos=None, rotation=0, edgecolor=None):
    edgecolor = edgecolor or color
    ax.add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, facecolor=color, edgecolor=edgecolor, lw=2))
    tx, ty = text_pos if text_pos else ((xmin + xmax) / 2, (ymin + ymax) / 2)
    ax.text(tx, ty, text, fontsize=14, color=edgecolor, rotation=rotation, ha='center', va='center')

#add_box(5e2, 2e3, 1.8, 10, 'xkcd:light grey', r'Reactor $\nu$', text_pos=(1.5e6, 4), edgecolor='xkcd:gray')
#add_box(5e4, 3e5, 1.8, 10, 'xkcd:light grey', r'', edgecolor='xkcd:gray')
add_box(3e6, 1.2e7, 1e4, 1e6, 'xkcd:peach', r'Atmospheric $\nu$', text_pos=(6.3e6, 1.2e5), rotation=90, edgecolor='xkcd:orange')

add_box(1.3e6, 1.3e6*1.05, 100, 5e3, 'xkcd:red', r'DUNE', text_pos=(1.3e6, 1.3e4), edgecolor='xkcd:red', rotation=90)
add_box(810e3, 810e3*1.04, 0.5e3, 5e3, 'xkcd:light orange', r'NO$\nu$A', text_pos=(8e5, 1.2e4), rotation=90)
add_box(250e3, 250e3*1.05, 100, 1.5e3, 'xkcd:green', r'T2K / Hyper-K', text_pos=(2.5e5, 9e3), rotation=90)


#add_box(110, 600, 100, 2e3, 'xkcd:sky blue', r'SBN', text_pos=(5e2, 250), edgecolor='xkcd:blue')
add_box(110, 110*1.05, 100, 2e3, 'xkcd:orange', r'SBND', text_pos=(110, 5e3), rotation=90)
add_box(470, 470*1.05, 100, 2e3, 'xkcd:blue', r'MicroBooNE', text_pos=(600 * 0.6, 1e4), rotation=90)
add_box(540, 540*1.05, 100, 2e3, 'xkcd:pink', r'MiniBooNE', text_pos=(600, 9e3), rotation=90)
add_box(600, 600*1.05, 100, 2e3, 'xkcd:green', r'ICARUS', text_pos=(600 * 1.6, 6.5e3), rotation=90)


add_box(30, 30*1.05, 20, 60, 'xkcd:light blue', r'LSND', text_pos=(30, 150), rotation=90)
add_box(24, 24*1.05, 10, 60, 'xkcd:lavender', r'JSNS$^2$', text_pos=(80, 30), rotation=90)
add_box(48, 48*1.05, 10, 60, 'xkcd:lavender', r'')

add_box(6.7, 9.2, 1.8, 10, 'xkcd:neon green', r'PROSPECT', text_pos=(8, 40), rotation=90)

add_box(500, 500*1.05, 1.8, 10, 'xkcd:hot pink', r'Daya Bay', text_pos=(800, 15))
add_box(1650, 1650*1.05, 1.8, 10, 'xkcd:hot pink', "")

add_box(140e3, 210e3, 1.8, 10, 'xkcd:aqua', r'KamLAND', text_pos=(8e5, 7))

add_box(52.5e3, 52.5e3*1.05, 1.8, 10, 'xkcd:neon purple', r'JUNO', text_pos=(2.5e4, 7))


plt.xscale('log')
plt.yscale('log')
plt.xlabel('L (meters)')
plt.ylabel('E (MeV)')
ax.set_xticks([1, 1e1, 1e2, 1e3, 1e4, 1e5, 1e6, 1e7, 1e8])
plt.xlim(1, 1e8)
plt.ylim(1, 1e6)
plt.show()
```
