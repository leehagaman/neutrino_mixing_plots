```python
import numpy as np
import matplotlib.pyplot as plt

from nufit_params import *
```

```python
m1_NO = 0
m2_NO = m1_NO + delta2_m21
m3_NO = m1_NO + delta2_m3l

m3_IO = 0
m2_IO = m3_IO - delta2_m3l_inv
m1_IO = m2_IO - delta2_m21_inv
```

```python
# alpha and beta set to zero here, Majorana phases should not matter for this plot
NO_PMNS = get_PMNS(theta12, theta23, theta13, deltaCP, 0, 0)
IO_PMNS = get_PMNS(theta12_inv, theta23_inv, theta13_inv, deltaCP_inv, 0, 0)

m1_NO_frac_e = np.abs(NO_PMNS[0][0])**2
m1_NO_frac_mu = np.abs(NO_PMNS[1][0])**2
m1_NO_frac_tau = np.abs(NO_PMNS[2][0])**2

m2_NO_frac_e = np.abs(NO_PMNS[0][1])**2
m2_NO_frac_mu = np.abs(NO_PMNS[1][1])**2
m2_NO_frac_tau = np.abs(NO_PMNS[2][1])**2

m3_NO_frac_e = np.abs(NO_PMNS[0][2])**2
m3_NO_frac_mu = np.abs(NO_PMNS[1][2])**2
m3_NO_frac_tau = np.abs(NO_PMNS[2][2])**2

m1_IO_frac_e = np.abs(IO_PMNS[0][0])**2
m1_IO_frac_mu = np.abs(IO_PMNS[1][0])**2
m1_IO_frac_tau = np.abs(IO_PMNS[2][0])**2

m2_IO_frac_e = np.abs(IO_PMNS[0][1])**2
m2_IO_frac_mu = np.abs(IO_PMNS[1][1])**2
m2_IO_frac_tau = np.abs(IO_PMNS[2][1])**2

m3_IO_frac_e = np.abs(IO_PMNS[0][2])**2
m3_IO_frac_mu = np.abs(IO_PMNS[1][2])**2
m3_IO_frac_tau = np.abs(IO_PMNS[2][2])**2

```

```python
print("NO m1 fractions, according to U_alphai^2:", m1_NO_frac_e, m1_NO_frac_mu, m1_NO_frac_tau)
print("NO m2 fractions, according to U_alphai^2:", m2_NO_frac_e, m2_NO_frac_mu, m2_NO_frac_tau)
print("NO m3 fractions, according to U_alphai^2:", m3_NO_frac_e, m3_NO_frac_mu, m3_NO_frac_tau)

print("\nIO m1 fractions, according to U_alphai^2:", m1_IO_frac_e, m1_IO_frac_mu, m1_IO_frac_tau)
print("IO m2 fractions, according to U_alphai^2:", m2_IO_frac_e, m2_IO_frac_mu, m2_IO_frac_tau)
print("IO m3 fractions, according to U_alphai^2:", m3_IO_frac_e, m3_IO_frac_mu, m3_IO_frac_tau)
```

```python
plt.rcParams.update({'font.size': 20})

fig = plt.figure(figsize=(10, 10))
ax = fig.gca()

rectangle_width = 0.8
rectangle_height = 0.00005

def make_rectangle(xmin, width, y, color):
    ax.add_patch(plt.Rectangle((xmin, y - rectangle_height/2), width, rectangle_height, color=color))

ax.text(-0.1, m1_NO, r"$m_1$", ha='center', va='center', fontsize=20)
make_rectangle(0, m1_NO_frac_e, m1_NO, 'tab:blue')
make_rectangle(m1_NO_frac_e, m1_NO_frac_mu, m1_NO, 'tab:orange')
make_rectangle(m1_NO_frac_e + m1_NO_frac_mu, m1_NO_frac_tau, m1_NO, 'tab:green')

ax.text(-0.1, m2_NO, r"$m_2$", ha='center', va='center', fontsize=20)
make_rectangle(0, m2_NO_frac_e, m2_NO, 'tab:blue')
make_rectangle(m2_NO_frac_e, m2_NO_frac_mu, m2_NO, 'tab:orange')
make_rectangle(m2_NO_frac_e + m2_NO_frac_mu, m2_NO_frac_tau, m2_NO, 'tab:green')

ax.text(-0.1, m3_NO, r"$m_3$", ha='center', va='center', fontsize=20)
make_rectangle(0, m3_NO_frac_e, m3_NO, 'tab:blue')
make_rectangle(m3_NO_frac_e, m3_NO_frac_mu, m3_NO, 'tab:orange')
make_rectangle(m3_NO_frac_e + m3_NO_frac_mu, m3_NO_frac_tau, m3_NO, 'tab:green')

IO_offset = 1.4

ax.text(IO_offset - 0.1, m1_IO, r"$m_1$", ha='center', va='center', fontsize=20)
make_rectangle(IO_offset, m1_IO_frac_e, m1_IO, 'tab:blue')
make_rectangle(IO_offset + m1_IO_frac_e, m1_IO_frac_mu, m1_IO, 'tab:orange')
make_rectangle(IO_offset + m1_IO_frac_e + m1_IO_frac_mu, m1_IO_frac_tau, m1_IO, 'tab:green')

ax.text(IO_offset - 0.1, m2_IO, r"$m_2$", ha='center', va='center', fontsize=20)
make_rectangle(IO_offset, m2_IO_frac_e, m2_IO, 'tab:blue')
make_rectangle(IO_offset + m2_IO_frac_e, m2_IO_frac_mu, m2_IO, 'tab:orange')
make_rectangle(IO_offset + m2_IO_frac_e + m2_IO_frac_mu, m2_IO_frac_tau, m2_IO, 'tab:green')

ax.text(IO_offset - 0.1, m3_IO, r"$m_3$", ha='center', va='center', fontsize=20)
make_rectangle(IO_offset, m3_IO_frac_e, m3_IO, 'tab:blue')
make_rectangle(IO_offset + m3_IO_frac_e, m3_IO_frac_mu, m3_IO, 'tab:orange')
make_rectangle(IO_offset + m3_IO_frac_e + m3_IO_frac_mu, m3_IO_frac_tau, m3_IO, 'tab:green')

ax.text(0.5, 0.0028, "Normal Ordering", ha='center', va='center', fontsize=20)
ax.text(0.5 + IO_offset, 0.0028, "Inverted Ordering", ha='center', va='center', fontsize=20)

plt.scatter([], [], c="tab:blue", label=r"$\nu_e$")
plt.scatter([], [], c="tab:orange", label=r"$\nu_\mu$")
plt.scatter([], [], c="tab:green", label=r"$\nu_\tau$")

# remove box from legend
plt.legend(loc="center", fontsize=30, frameon=False)


plt.ylim(-0.0001, 0.003)
plt.xlim(-0.3, IO_offset + 1.1)

ax.get_xaxis().set_visible(False)

plt.ylabel("$m^2 - m_\mathrm{min}^2$ (eV$^2$)")

plt.tight_layout()

plt.savefig("plots/NO_IO_neutrino_masses.pdf")
```
